### Maximum snow_depth in the data

```python
max_snow_depth = data.groupby('country')['snow_depth'].max().reset_index()
max_snow_depth.columns = ['country', 'max_snow_depth']
max_snow_depth
```

```python
plt.figure(figsize=(10, 6))
plt.bar(max_snow_depth['country'], max_snow_depth['max_snow_depth'], color='skyblue')
plt.title('Maximum Snow Depth for Each Country')
plt.xlabel('Country')
plt.ylabel('Max Snow Depth (cm)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### All Precipitation above 10 

```python
data[data['precipitation'] > 10]
```

```python

data['date'] = pd.to_datetime(data['date'])
precipitation_above_10 = data[data['precipitation'] > 10]
plt.figure(figsize=(10, 6))
plt.plot(precipitation_above_10['date'], precipitation_above_10['precipitation'], marker='o', linestyle='-', color='b')
plt.title('Precipitation Above 10')
plt.xlabel('Date')
plt.ylabel('Precipitation')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Rename tavg, tmin and tmax to temp_avg, temp_min and temp_max

```python
data.rename(columns = {'tavg' : 'temp_avg', 'tmin':'temp_min', 'tmax':'temp_max'}, inplace = True)
```

```python
data
```

### Mean temp_max

```python
data.temp_max.mean()
```

```python
data['date'] = pd.to_datetime(data['date'])
heatmap_data = data.pivot_table(index='date', columns='country', values='temp_max')

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".1f", linewidths=.5)
plt.title('Heat Map of Maximum Temperatures')
plt.xlabel('Country')
plt.ylabel('Date')
plt.tight_layout()
plt.show()
```

### All data where precipitation is above 15 and snow_depth is above 200

```python
data[(data['precipitation'] > 15) & (data['snow_depth'] > 200)]
```

```python
data['date'] = pd.to_datetime(data['date'])
filtered_data = data[(data['precipitation'] > 15) & (data['snow_depth'] > 200)]
plt.figure(figsize=(14, 8))
plt.scatter(filtered_data['date'], filtered_data['snow_depth'], c=filtered_data['precipitation'], cmap='viridis', alpha=0.7, edgecolors='w', s=100)
plt.title('Scatter Plot of Snow Depth and Precipitation')
plt.xlabel('Date')
plt.ylabel('Snow Depth (cm)')
plt.colorbar(label='Precipitation (mm)')
plt.tight_layout()
plt.show()
```
